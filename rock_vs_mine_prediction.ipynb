Rock vs mine prediction


```python
import pandas as pd # displaying data flame
import numpy as np  # generating array
from sklearn.model_selection import train_test_split # to be used in splitting the data
from sklearn.linear_model import LogisticRegression   # tthis is the model to be used
from sklearn.metrics import accuracy_score    #checking the accuracy of the model
```

data collection and data processing

```python
#Loading the dataset to a pandas dataframe
sonar_data=pd.read_csv('/content/drive/MyDrive/Copy of sonar data.csv')
sonar_data.head()
```

```python
#sonar_data.isnull().sum()
# number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe()  # describe statistical measures of the data
```

```python
# Check the columns of the DataFrame
print(sonar_data.columns)

# If you need to get a list of column names as integers, you can convert them
columns_as_integers = [int(col) for col in sonar_data.columns if col.isdigit()]

print(columns_as_integers)
```

```python
sonar_data['R'].value_counts()
```

m represents mine and r represents rock

```python
sonar_data.groupby('R').mean()
```

```python
# splitting the data and labels
x=sonar_data.drop(columns='R',axis=1)
y=sonar_data['R']
```

```python
print(x)
print(y)
```

Training and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x_train.shape,y_train.shape,x.shape,y.shape)
```

Model training  --->logistic regression model

```python
# training the logistic regression model using the training data
model=LogisticRegression()
model.fit(x_train,y_train)
```

model evaluation

```python
# accuracy on the training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('accuracy on training data',training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('accuracy ontest data:',test_data_accuracy)
```
